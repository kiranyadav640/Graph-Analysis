```python
def delete_nodes_in_graph():
    G=nx.Graph()
    G.add_edge('A','B')
    
    G.add_edge('B','C')
    G.add_edge('C','A')
    nx.draw(G,node_color='y',node_size=800,with_labels=True)
    G.remove_node('B')
    nx.draw(G,node_color='cyan',node_size=800,with_labels=True)
    
```

```python
delete_nodes_in_graph()
```

```python
def delete_specific_edge():
    G=nx.Graph()
    G.add_edge('A','B')
    
    G.add_edge('B','C')
    G.add_edge('C','A')
    nx.draw(G,node_color='y',node_size=800,with_labels=True)
    G.remove_edge('A','B')
    G.remove_edge('C','A')
    nx.draw(G,node_color='cyan',node_size=800,with_labels=True)
    
```

```python
def custom_node_size_and_node_color():
    G=nx.Graph()
    G.add_edge('A','B')
    
    G.add_edge('B','C')
    G.add_edge('C','A')
    nx.draw(G,node_color='y',node_size=800,with_labels=True)
    plt.show()
    
    nodelst=['A','B','C']
    colorlst=['c','m','y']
    nodesize=[1000,2000,3000]
    nx.draw(G,node_color=colorlst,node_size=nodesize,nodelist=nodelst,with_labels=True)
```

```python
custom_node_size_and_node_color()
```

```python
def custom_node_size_and_node_color():
    G=nx.Graph()
    G.add_edge('A','B')
    
    G.add_edge('B','C')
    G.add_edge('C','A')
    nx.draw(G,node_color='y',node_size=800,with_labels=True)
    plt.show()
    
    edgelst=[('A','B'),('B','C'),('C','A')]
    edgecolor=['c','y','m']
    colorlst=['c','m','y']
    nodesize=[1000,2000,3000]
    nx.draw(G,node_color=colorlst,edgelist=edgelst,edge_color=edgecolor,node_size=nodesize,with_labels=True)
```

```python
custom_node_size_and_node_color()
```
